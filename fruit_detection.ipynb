```python
import cv2
import os
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import joblib

dataset_path = "train"

label_map = {
    "apple": 0,
    "banana": 1,
    "orange": 2
}

features = []
labels = []

for fruit in os.listdir(dataset_path):
    fruit_folder = os.path.join(dataset_path, fruit)
    if fruit not in label_map:
        continue

    for img_name in os.listdir(fruit_folder):
        if not img_name.lower().endswith(('.jpg', '.png', '.jpeg')):
            continue

        img_path = os.path.join(fruit_folder, img_name)
        img = cv2.imread(img_path)

        if img is None:
            continue

        img = cv2.resize(img, (100, 100))
        mean_color = img.mean(axis=(0, 1))  # BGR color average

        features.append(mean_color)
        labels.append(label_map[fruit])

X = np.array(features)
y = np.array(labels)

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)

joblib.dump(model, "fruit_model.pkl")

print("Model trained and saved.")
```

```python
import cv2
import numpy as np
import joblib

model = joblib.load("fruit_model.pkl")

class_names = {
    0: "apple",
    1: "banana",
    2: "orange"
}

def detect_fruit(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (100, 100))

    mean_color = img.mean(axis=(0, 1))
    mean_color = mean_color.reshape(1, -1)

    prediction = model.predict(mean_color)[0]
    return class_names[prediction]
```

```python
result = detect_fruit("test/orange/orange_78.jpg")
print("Detected fruit:", result)
```
